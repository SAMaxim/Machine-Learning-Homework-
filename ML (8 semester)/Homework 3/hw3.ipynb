# HW 3 - NLP
## Генератор слоганов
В рамках данного задания необходимо сделать алгоритм (нейросеть), которая получает на вход имя, а на выходе выдает забавный слоган.
Обязательные условия:
- Формат выхода: "Имя" - "слоган с маленькой буквы".
- Работа на русском языке.
- Имя должно учитываться как seed.
- Слоган должен быть не короче 3 слов.
- Слоган должен генерироваться, а не выбираться из базы данных и не повторять примеры из базы данных.
- Слоган должен быть связным текстом без ошибок в словах.

Допущения:
- Можно не использовать пунктуацию.
- Время работы не ограничено.
- Работоспособность достаточно показать на 3-х именах (во время сдачи показать для одного примера).
- Обучающая выборка и сложность модели могут быть произвольными.
- Слоганы могут быть однообразными.

Критерий оценивания:
- Чем абсурднее и забавнее - тем лучше.

## Подсказки / варианты решения

1. Работайте с символами (буквами) (чтобы уменьшить словарь и уметь обрабатывать имена, которых алгоритм не видел).
2. При обучении приводите все к нижнему регистру и к начальной форме.
3. Для предобучения можно использовать [частушки](http://opentextnn.ru/old/museum/nn/aetnolog/folk/index.html@id=4424) / цитаты комиков / [другие генераторы слоганов](https://castlots.org/generator-sloganov/) (реализовать цикл с запросами на сайт). Также хорошим вариантом будут анекдоты, в которых встречаются имена.
4. Можно отдельно отбирать тексты, где встречаются имена и выносить их, как ключевое слово.
5. Дообучаться можно на [рекламных слоганах](https://yagla.ru/blog/marketing/99-samyh-izvestnyh-sloganov-v-rossii/).
6. При генерации слогана можно проверять наличие слова в словаре ([словать взять готовый для русского языка](https://ru.stackoverflow.com/questions/1157034/%D0%92%D1%81%D0%B5-%D1%80%D1%83%D1%81%D1%81%D0%BA%D0%B8%D0%B5-%D1%81%D0%BB%D0%BE%D0%B2%D0%B0)). Если слова нет, можно не выдавая пользователю результат попробовать еще раз.
7. Приводите слова в слогане к правильной форме перед выдачей пользователю, с помощью, например, [pymorphy2](https://pymorphy2.readthedocs.io/en/stable/user/grammemes.html), [pymorphy3](https://pypi.org/project/pymorphy3/), [pymystem3](https://nlpub.ru/Mystem).
8. Начать работу рекомендуется с LSTM в несколько слоев.
